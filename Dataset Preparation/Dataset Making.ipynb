```python
import pandas as pd
```

```python
from datasets import load_dataset
dataset = load_dataset("lavita/ChatDoctor-HealthCareMagic-100k", split = "train")
```

```python
df = dataset.to_pandas()
```

```python
df
```

```python
def formatting_prompts_func(row):
  return f"""<|start_header_id|>system<|end_header_id|>{row["instruction"]}<|eot_id|><|start_header_id|>user<|end_header_id|> This is the question: {row["input"]}<|eot_id|><|start_header_id|>assistant<|end_header_id|>{row["output"]}<|eot_id|>"""
```

```python
df = pd.DataFrame(df)
```

```python
df['prompt'] = df.apply(formatting_prompts_func, axis=1)
```

```python
import datasets
dataset = datasets.Dataset.from_pandas(df)
dataset.push_to_hub("srikar-v05/LLaMA3-Instruct-Medical",token="YOUR_HUGGINGFACE_TOKEN")
```
